```python
from ultralytics import YOLO
import cv2

keys = {0: "fish", 1: "human"}
model = YOLO("./best.pt")


def predict(img):
    results = model.predict(img)

    for id, result in enumerate(results):
        try:
            cls = keys[int(result.boxes.cls[id].item())]
            boxes = result.boxes
            for idx in range(len(boxes.xyxy)):
                x1, y1, x2, y2 = boxes.xyxy[idx]
                x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
                color = (0, 255, 0)
                if cls == "human":
                    color = (255, 0, 0)
                img = cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
                img = cv2.putText(img ,cls ,((x1+x2)//2, y2),cv2.FONT_HERSHEY_COMPLEX, 1, color, 2, cv2.LINE_AA)

        except:
            continue

    return img
```

```python
camera = cv2.VideoCapture(0)

while True:
    _, frame = camera.read()
    try:
        frame = predict(frame)
    
        cv2.imshow('camera', frame)
    except:
        break
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
camera.release()
cv2.destroyAllWindows()
```
